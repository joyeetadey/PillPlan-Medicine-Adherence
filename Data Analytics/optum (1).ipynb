```python
#maintaing a list showing if the medical condition is ever stopped or not
list_my=[]
for x in df_con['STOP']:
    if(x==0):
        list_my.append(0)
    else:
        list_my.append(1)
```

```python

list2=[]
for i in range(8376):
    list2.append(1)
df_con['SCORE']=list_my
df_con['COUNT']=list2
```

```python
df_con.head()
```

```python
# df_score=pd.pivot_table(df_con, index=['PATIENT'],values=['SCORE','COUNT'],aggfunc=np.sum)
# df_score.head()


#maintaing a score file for predicting if the patient is adherent or not
df_score = df_con.groupby(['PATIENT'])['SCORE', 'COUNT'].sum().reset_index()

df_score
```

```python
df_score
```

```python
df_score['weight']=df_score['SCORE']/df_score['COUNT']
```

```python
df_score.head()
```

```python
#if the weight  of the particular patient id more than 60% we classify it in the category of adherence 
list3=[]
for x in df_score['weight']:
    if(x>=0.6):
        list3.append(1)
    else:
        list3.append(-1)
        
```

```python
df_score['adh']=list3
```

```python
from datetime import datetime
def duration(date):
    date = date[:-10] 
    if(len(date)>9):
      date = datetime.datetime.strptime(date, '%Y-%m-%d')
      start_date = datetime.date(date.year, date.month, date.day)
      end_date = datetime.date.today()
      time_difference = end_date - start_date
      age = time_difference.days
      return age
    else:
      return 0
```

```python
r3['DURATION_1'] = r3['MED_START'].apply(lambda x:duration(x))
r3['DURATION_2'] = r3['MED_STOP'].apply(lambda x:duration(str(x)))
r3['DURATION'] = r3['DURATION_1']-r3['DURATION_2']
r3 = r3.drop(['DURATION_1','DURATION_2'],axis=1)
r3.head()
```
